## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
df.head()
```

```python
# your code goes here

import matplotlib.pyplot as plt

QUERY = """
SELECT ConvertedComp
FROM master
"""
df=pd.read_sql_query(QUERY,conn)

plt.figure(figsize=(10, 6))
plt.hist(df, bins=5, edgecolor='black')  # Adjust bins as needed
plt.title('Histogram of Converted Compensation')
plt.xlabel('Converted Compensation (USD)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here

QUERY = """
SELECT Age, WorkWeekHrs
FROM master
"""

# Execute the query and read data into a DataFrame
df = pd.read_sql_query(QUERY, conn)


# Plotting a scatter plot of 'Age' versus 'WorkWeekHrs'
plt.figure(figsize=(10, 6))
plt.scatter(df['Age'], df['WorkWeekHrs'], alpha=0.5)
plt.title('Scatter Plot of Age vs WorkWeekHrs')
plt.xlabel('Age')
plt.ylabel('Work Week Hours')
plt.grid(True)
plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here

QUERY = """
SELECT WorkWeekHrs, CodeRevHrs, Age
FROM master
"""

df = pd.read_sql_query(QUERY, conn)

# Plotting a bubble plot of 'WorkWeekHrs' against 'CodeRevHrs', with bubble size based on 'Age'
plt.figure(figsize=(10, 6))
plt.scatter(df['WorkWeekHrs'], df['CodeRevHrs'], s=df['Age']*10, alpha=0.5)
plt.title('Bubble Plot of WorkWeekHrs vs CodeRevHrs')
plt.xlabel('WorkWeekHrs')
plt.ylabel('CodeRevHrs')
plt.grid(True)
plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here

QUERY = """
SELECT *
FROM DatabaseDesireNextYear
"""

df = pd.read_sql_query(QUERY, conn)
top_databases = df['DatabaseDesireNextYear'].value_counts().head(5)
labels = top_databases.index.tolist()
sizes = top_databases.values.tolist()

# Plotting the pie chart
plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Top 5 Databases Respondents Wish to Learn Next Year')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here

QUERY = """
SELECT Age, WorkWeekHrs, CodeRevHrs
FROM master
WHERE Age BETWEEN 30 AND 35
"""

df = pd.read_sql_query(QUERY, conn)

median_values = df.groupby('Age')[['WorkWeekHrs', 'CodeRevHrs']].median()

# Plotting the stacked bar chart
median_values.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Median WorkWeekHrs and CodeRevHrs for Age Group 30-35')
plt.xlabel('Age')
plt.ylabel('Median Hours')
plt.xticks(rotation=0)
plt.legend(loc='upper right')
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here

QUERY = """
SELECT Age, ConvertedComp
FROM master
WHERE Age BETWEEN 45 AND 60
"""

# Execute the query and read data into a DataFrame
df = pd.read_sql_query(QUERY, conn)

median_values = df.groupby('Age')['ConvertedComp'].median()

# Plotting the line plot of median ConvertedComp against age
plt.figure(figsize=(10, 6))
plt.plot(median_values.index, median_values.values, marker='o', linestyle='-', color='b')
plt.title('Median Converted Compensation for Ages 45-60')
plt.xlabel('Age')
plt.ylabel('Median Converted Compensation (USD)')
plt.grid(True)
plt.xticks(median_values.index)
plt.show()
```

```python
# your code goes here

QUERY = """
SELECT Age, ConvertedComp
FROM master
WHERE Age BETWEEN 25 AND 30
"""

# Execute the query and read data into a DataFrame
df = pd.read_sql_query(QUERY, conn)

median_values = df.groupby('Age')['ConvertedComp'].median()

# Plotting the line plot of median ConvertedComp against age
plt.figure(figsize=(10, 6))
plt.plot(median_values.index, median_values.values, marker='o', linestyle='-', color='b')
plt.title('Median Converted Compensation for Ages 25-30')
plt.xlabel('Age')
plt.ylabel('Median Converted Compensation (USD)')
plt.grid(True)
plt.xticks(median_values.index)
plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here
QUERY = """
SELECT MainBranch
FROM master
"""

# Execute the query and read data into a DataFrame
df = pd.read_sql_query(QUERY, conn)
branch_counts = df['MainBranch'].value_counts()

# Plotting the horizontal bar chart
plt.figure(figsize=(10, 6))
branch_counts.plot(kind='barh', color='skyblue')
plt.title('Distribution of Main Branches')
plt.xlabel('Count')
plt.ylabel('Main Branch')
plt.gca().invert_yaxis()  # Invert y-axis to have the highest count at the top
plt.show()
```

```python

# Define your SQL query to retrieve data from 'LanguageDesireNextYear' table
QUERY = """
SELECT LanguageDesireNextYear, COUNT(*) AS NumRespondents
FROM LanguageDesireNextYear
GROUP BY LanguageDesireNextYear
ORDER BY NumRespondents DESC
"""

# Execute the query and read data into a DataFrame
df = pd.read_sql_query(QUERY, conn)


# Reset index for rank calculation
df.reset_index(drop=True, inplace=True)

# Find the rank of Python
python_rank = df[df['LanguageDesireNextYear'] == 'Python'].index.tolist()[0] + 1

print(f"The rank of Python among the most popular languages respondents wish to learn next year is: {python_rank}")
```

```python


# Define your SQL query to retrieve data from 'LanguageWorkedWith' table
QUERY = """
SELECT COUNT(*) AS NumRespondents
FROM LanguageWorkedWith
WHERE LanguageWorkedWith LIKE '%SQL%'
"""

# Execute the query and read data into a DataFrame
df = pd.read_sql_query(QUERY, conn)

# Extract the number of respondents who work with SQL
num_respondents_sql = df['NumRespondents'].iloc[0]

print(f"The number of respondents who currently work with SQL is: {num_respondents_sql}")
```

```python

QUERY = """
SELECT COUNT(*) AS NumRespondents
FROM DatabaseWorkedWith
WHERE DatabaseWorkedWith = 'MySQL'
AND DatabaseWorkedWith NOT IN (
    SELECT DatabaseWorkedWith
    FROM DatabaseWorkedWith
    WHERE DatabaseWorkedWith != 'MySQL'
)
"""

# Execute the query and read data into a DataFrame
df = pd.read_sql_query(QUERY, conn)

# Extract the number of respondents who work exclusively with MySQL
num_respondents_mysql_only = df['NumRespondents'].iloc[0]

print(f"The number of respondents who work exclusively with MySQL is: {num_respondents_mysql_only}")
```

```python
QUERY = """
SELECT DevType, COUNT(*) AS RespondentCount
FROM DevType
GROUP BY DevType
ORDER BY RespondentCount DESC
LIMIT 1;
"""
cursor = conn.cursor()
cursor.execute(QUERY)

# Fetch the result
result = cursor.fetchone()

# Print the result
if result:
    dev_type, respondent_count = result
    print(f"The majority of survey responders are from '{dev_type}' with {respondent_count} respondents.")
else:
    print("No data found.")
```
